Reproduce this bar graph.

```python
courses = ['PF', 'DL', 'CV']
y_pos = [2, 0, 1]
scores = [70, 60, 40]

data = [courses, y_pos, scores]

import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data)
df = df.T
df.columns=['courses', 'y', 'scores']

df.iloc[0], df.iloc[2] = df.iloc[2].copy(), df.iloc[0].copy()
df.iloc[0], df.iloc[1] = df.iloc[1].copy(), df.iloc[0].copy()

plt.figure(figsize=(5,3))
plt.bar(df['courses'], df['scores'], color='seagreen',)
plt.ylabel('scores')
plt.title('Gradebook of Nida')
plt.ylim(0,100)
plt.show()
```

Reproduce this figure.  

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit

noise = np.random.uniform(-20, 20, 100)
x = np.linspace(0, 100, 100) + noise
y = np.linspace(0, 50, 100) + noise

plt.figure(figsize=(5, 3))
plt.scatter(x, y, label='Scatter Plot', color='purple')

b, m = polyfit(x, y, 1)
y_pred = m * x + b

# Plot the line of best fit
plt.plot(x, y_pred, color='black', label='Line of Best Fit')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Line of Best Fit')

plt.legend()
plt.show()
```
